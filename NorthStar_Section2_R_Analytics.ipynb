# Load libraries


```python
%%R

install.packages("ggplot2")
install.packages("dplyr")
install.packages("tidyr")
install.packages("corrplot")

library(ggplot2)
library(dplyr)
library(tidyr)
library(corrplot)
```

# Load datasets

```python
%%R

customers <- read.csv("customers.csv")

orders <- read.csv("orders.csv")

drivers <- read.csv("drivers.csv")

vehicles <- read.csv("vehicles.csv")

complaints <- read.csv("complaints.csv")

app_events <- read.csv("app_events.csv")

incidents <- read.csv("incidents.csv")

hubs <- read.csv("hubs.csv")

deliveries <- read.csv("deliveries.csv")
```

# Analyse 1 - Distribution of order values

```python
%%R

ggplot(orders, aes(order_value)) +

geom_histogram(bins=30) +

labs(
title="Distribution of Order Values",
x="Order Value",
y="Frequency"
)
```

# Analyse 2 -  Service demand volume

```python
%%R

ggplot(orders, aes(service_type)) +

geom_bar() +

labs(
title="Demand by Service Type",
x="Service Type",
y="Number of Orders"
)
```

# Analyse 3 -  Complaint severity patterns

```python
%%R

ggplot(complaints, aes(severity)) +

geom_bar() +

labs(
title="Complaint Severity Distribution",
x="Severity Level",
y="Total Complaints"
)
```

# Analyse 4 - Driver rating distribution

```python
%%R

ggplot(drivers, aes(driver_rating)) +

geom_histogram(bins=20) +

labs(
title="Driver Rating Distribution",
x="Driver Rating",
y="Frequency"
)
```

# Analyse 5 -  Average order value by service type

```python
%%R

avg_value <- orders %>%

group_by(service_type) %>%

summarise(avg_value = mean(order_value, na.rm=TRUE))

ggplot(avg_value, aes(service_type, avg_value)) +

geom_bar(stat="identity") +

labs(
title="Average Order Value by Service",
x="Service Type",
y="Average Order Value"
)
```

# Analyse 6 -  Complaint resolution time

```python
%%R

ggplot(complaints, aes(resolution_days)) +

geom_histogram(bins=20) +

labs(
title="Complaint Resolution Time",
x="Days",
y="Frequency"
)
```

# Analyse 7 -  App performance latency

```python
%%R

ggplot(app_events, aes(api_latency_ms)) +

geom_histogram(bins=30) +

labs(
title="App API Latency",
x="Latency (ms)",
y="Frequency"
)
```

# Analyse 8 - Incident severity frequency

```python
%%R

ggplot(incidents, aes(severity)) +

geom_bar() +

labs(
title="Incident Severity Frequency",
x="Severity",
y="Count"
)
```
